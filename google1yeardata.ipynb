```python
import pandas as pd 
import numpy as np
import yfinance as yf
from yahoofinancials import YahooFinancials
import plotly.graph_objs as go
```

```python
data = yf.download(tickers='GOOG', period='365d', interval='1h')
```

```python
data
```

```python
fig = go.Figure()
```

```python
fig.add_trace(go.Candlestick(x=data.index,
                open=data['Open'],
                high=data['High'],
                low=data['Low'],
                close=data['Close'], name = 'market data'))
```
